## Predict Hospital Admitted COVID Pateitns given COVID-19 CASES

## Feature Engineering

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
import plotly.graph_objects as go
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

## Model Training


```python

# Load the data
covid_data = pd.read_csv('../datasets/filtered_datasets/cases_state.csv', parse_dates=['date'])
hospital_data = pd.read_csv('../datasets/filtered_datasets/hospital.csv', parse_dates=['date'])

# Merge the data on the date column
data = pd.merge(covid_data, hospital_data, on=['date', 'state'])

# Sort the data by date
data = data.sort_values(by='date')
```

## Prediction and Evaluation using rmse, mse, mae


```python
# List of states
states = data['state'].unique()

# Dictionary to store results for each state
results = {}

for state in states:
    # Filter data for the current state
    state_data = data[data['state'] == state]
    
    # Select the target variable
    target = 'admitted_covid'
    y = state_data.set_index('date')[target]
    
    # Split the data into training and testing sets
    train_end = '2022-06-30'
    test_start = '2022-07-01'
    test_end = '2022-07-31'
    y_train = y[:train_end]
    y_test = y[test_start:test_end]
    
    # Fit the ARIMA model
    model = ARIMA(y_train, order=(15, 2, 1))  
    model_fit = model.fit()
    
    # Make predictions for the entire test period
    predictions = model_fit.predict(start=test_start, end=test_end, typ='levels')
    
    # Ensure y_test and predictions have the same index
    y_test = y_test.loc[predictions.index]
    
    # Evaluate the model
    rmse = np.sqrt(mean_squared_error(y_test, predictions))
    mse = mean_squared_error(y_test, predictions)
    mae = mean_absolute_error(y_test, predictions)
    
    print(f'State: {state}')
    print(f'Root Mean Squared Error: {rmse}')
    print(f'Mean Squared Error: {mse}')
    print(f'Mean Absolute Error: {mae}\n')
    
    # Store results
    results[state] = {
        'model': model_fit,
        'predictions': predictions,
        'rmse': rmse,
        'mse': mse,
        'mae': mae
    }
```

## Plot actual vs predicted values


```python
# Visualization: Actual vs Predicted values for a specific state
state_to_plot = 'W.P. Putrajaya'
state_data = data[data['state'] == state_to_plot]
y = state_data.set_index('date')[target]

# Combine the actual and predicted values into a single DataFrame for easier plotting
plot_data = pd.DataFrame({
    'date': y.index,
    'actual': y,
    'predicted': np.nan
})
plot_data.loc[results[state_to_plot]['predictions'].index, 'predicted'] = results[state_to_plot]['predictions']

# Plot the results
fig = go.Figure()
fig.add_trace(go.Scatter(x=plot_data['date'], y=plot_data['actual'], mode='lines', name='Actual', line=dict(color='blue')))
fig.add_trace(go.Scatter(x=plot_data['date'], y=plot_data['predicted'], mode='lines', name='Predicted', line=dict(color='red', dash='dash')))
fig.update_layout(xaxis_title='Date', yaxis_title='Admitted COVID Patients', title=f'Actual vs Predicted Admitted COVID Patients in {state_to_plot}')
fig.show()
```
